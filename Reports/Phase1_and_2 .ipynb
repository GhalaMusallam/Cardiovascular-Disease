```python
#Detect and remove outliers (remove noise)

outlier_threshold = 1.5

def detect_outliers(column_data):
    # Calculate the first and third quartile
    q1 = np.percentile(column_data, 25)
    q3 = np.percentile(column_data, 75)

    iqr = q3 - q1

    upper_bound = q3 + outlier_threshold * iqr
    lower_bound = q1 - outlier_threshold * iqr

    outliers = (column_data > upper_bound) | (column_data < lower_bound)

    return outliers

numeric_columns = df.select_dtypes(include=[np.number]).columns

cleaned_data = df.copy()

for column in numeric_columns:
    outliers = detect_outliers(cleaned_data[column])

    cleaned_data = cleaned_data[~outliers]

print(f"Original data had {len(df)} rows")
print(f"Cleaned data has {len(cleaned_data)} rows after removing outliers")

```

```python
#Discretization
column_to_discretize = 'age'
num_bins = 3

df['discretized_' + column_to_discretize] = pd.cut(df[column_to_discretize], bins=num_bins, labels=False)

print("Original DataFrame:")
print(df[['age', 'discretized_' + column_to_discretize]])
```

The output shows the original 'age' values and the corresponding discretized values (0, 1, or 2) for each row, based on the bin that the age value falls into.

The discretization process transforms the continuous 'age' values into a set of categorical bins, which can be useful for certain types of data analysis or modeling tasks where the underlying variable needs to be treated as a categorical feature rather than a continuous one.

```python
#Correlation-based Feature Selection (Filter Method)
X = df.drop(columns=['weight'])
y = df['weight']
selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]

print("Selected Features:", selected_features)
```

```python
#Variance Threshold (Filter Method)
X = df.drop(columns=['height'])
y = df['height']

columns = X.columns
print("Original Dataset:")
print(X.head())

selector = VarianceThreshold(threshold=0.2)
X_new = selector.fit_transform(X)

selected_features = [columns[i] for i in range(X.shape[1]) if selector.get_support()[i]]
print("\nSelected Features:")
print(X[selected_features].head())
```

```python
#Recursive Feature Elimination(Wrapper method)
x = df[['height', 'weight', 'age']]
y = df['cardio']
model = LogisticRegression()

rfe = RFE(model, n_features_to_select=1)
x_new = rfe.fit_transform(x, y)

selected_features = [i for i, mask in enumerate(rfe.support_) if mask]

print("Selected Features:", selected_features)
```

```python
#L1 regulation (Embedded method)
X = df.drop(columns=['ap_hi'])
y = df['ap_hi']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = Lasso(alpha=0.1, max_iter=5000)  # Increased max_iter
model.fit(X_scaled, y)

selected_features = X.columns[model.coef_ != 0]
print("Selected Features:", selected_features)
```

---------------------------------------------------------------
After implementing each feature selection method we can see that the methods L1 Regulation(Embedded method) and Variance Threshold(Filter method) selected more features compared to the others.

```python
#Normalization
columns_to_normalize = ['age', 'height']
data_to_normalize = df[columns_to_normalize]

# Min-Max Scaling
minmax_scaler = MinMaxScaler()
normalized_data_minmax = minmax_scaler.fit_transform(data_to_normalize)
df[columns_to_normalize] = normalized_data_minmax
print("Min-Max scaled data (age and height columns):")
print(df)

# Z-score Normalization
zscore_scaler = StandardScaler()
normalized_data_zscore = zscore_scaler.fit_transform(data_to_normalize)
df[columns_to_normalize] = normalized_data_zscore
print("Z-score normalized data:")
print(df)

# Decimal Scaling Normalization
for column in columns_to_normalize:
    max_abs_value = df[column].abs().max()
    df[column] = df[column] / (10 ** len(str(int(max_abs_value))))
print("DataFrame after Decimal Scaling Normalization:")
print(df)
```

**Min-Max Scaled Data:**

-This shows the 'age' and 'height' columns after applying min-max scaling normalization.

-Min-max scaling normalizes the values to a range between 0 and 1.

-For example, the 'age' values range from 0.0 to 99999.0, which have been scaled to a range of 0.0 to 1.0.

-Similarly, the 'height' values have been scaled to a range of 0.579487 to 0.654103.

**Z-score Normalized Data:**

-This shows the 'age' and 'height' columns after applying Z-score normalization.

-Z-score normalization transforms the data to have a mean of 0 and a standard deviation of 1.

-The 'age' values now range from around -0.436062 to 99993.0 in Z-score units.

-The 'height' values range from around -1.018168 to 2.270477 in Z-score units.

**DataFrames after Decimal Scaling Normalization:**

-This shows the entire DataFrame after applying decimal scaling normalization to all columns.

-Decimal scaling divides each value by a power of 10, based on the maximum absolute value in the column.

-This ensures that all values are scaled between 0 and 1.

-For example, the 'age' column is divided by 99999.0, the maximum value, to be scaled between 0.0 and 1.0.

-Similarly, the other columns like 'height', 'weight', 'ap_hi', and 'ap_lo' are all scaled using decimal scaling.

The normalization techniques are used to ensure that the features in the dataset have a similar scale, which can be important for many machine learning algorithms. This helps to prevent features with larger numerical ranges from dominating the learning process.
